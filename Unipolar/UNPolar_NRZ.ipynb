```python
# This code generates a Unipolar NRZ (Non-Return-to-Zero) line coding representation for a binary input sequence provided by the user.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def unipolarNRZ(X):
    # Loop through each bit in the input sequence X
    n = 1
    L = len(X)
    X.append(1)
    while n <= len(X)-1:
        # Generate a smoother time array within the range of each bit
        t = np.linspace(n-1, n, 1000)
        # Determine the signal level based on current and next bits for Unipolar NRZ line coding
        if X[n-1] == 0:
            if X[n] == 0:
                y = (t > n).astype(int)
            else:
                y = (t == n).astype(int)
        else:
            if X[n] == 0:
                y = (t < n).astype(int) - 0*(t == n).astype(int)
            else:
                y = (t < n).astype(int) + 1*(t == n).astype(int)

        # Plot the signal waveform for each bit
        plt.plot(t, y, linewidth=2.5)
        plt.title('Unipolar NRZ line coding')
        plt.grid(True)
        plt.axis([0, len(X)-1, -1.5, 1.5])
        n += 1
```

```python
# The main function prompts the user to enter a binary input sequence, converts it to a list of integers,
# and then plots the Unipolar NRZ line coding representation of the input sequence using Matplotlib.



def main():
    print("\n\n***************************************************************************************")
    X = input("---> Enter binary input sequence (e.g., 1011100): ")
    X = [int(bit) for bit in X]
    print("\n\n")
    # Create a new figure and plot the polar NRZ line coding for the input sequence X
    plt.figure()
    plt.subplot(2, 1, 1)
    unipolarNRZ(X)
    plt.suptitle('Input: ' + ' '.join(map(str, X)))
    plt.show()
    plt.show()

if __name__ == "__main__":
    main()
```
