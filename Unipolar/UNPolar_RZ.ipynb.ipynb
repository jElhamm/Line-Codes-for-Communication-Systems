```python
# This code defines a function unipolar_rz_plot that generates a Unipolar Return-to-Zero (RZ) line coding plot for binary input data.
# The function extends each bit in the input binary data by 1000 samples to visualize the encoding.
# The main function main allows the user to input binary data as a string, converts it to a list of integers,
# and then plots the Unipolar RZ line coding for the input data.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def unipolarRZ(X):
    encoded_signal = []                                                     # Initialize an empty list to store the encoded signal
    for i in range(len(X)):                                                 # Iterate through each bit in the input data X
        encoded_signal.extend([X[i]] * 1000)                                # Extend each bit value for 1000 samples

    t = np.linspace(0, len(X), len(encoded_signal), endpoint=False)         # Generate the time axis for the plot
    plt.figure(figsize=(8, 4))                                              # Create a plot to visualize the Unipolar RZ line coding
    plt.plot(t, encoded_signal, linewidth=2)
    plt.title('Unipolar RZ line coding')                                    # Set plot title, x-axis label, y-axis label, and grid
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.ylim(-0.5, 1.5)                                                     # Set the y-axis limits for better visualization
```

```python
# The main function prompts the user to enter a binary input sequence, converts it to a list of integers,
# and then plots the unipolar RZ line coding representation of the input sequence using Matplotlib.



def main():
    print("\n\n***************************************************************************************")
    X_str = input("---> Enter the binary data (e.g., 1011100): ")
    X = [int(x) for x in X_str]
    print("\n\n")
    unipolarRZ(X)


if __name__ == "__main__":
    main()
```
