```python
# This code defines a function polarNRZ(X) that implements Polar NRZ (Non-Return-to-Zero) line coding for a given binary input sequence.
# It encodes the input sequence into polar representations and plots the Polar RZ line coding representation using Matplotlib.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polarNRZ(X):
    n = 1
    L = len(X)
    X.append(1)                               # Append a terminate bit
    t = np.arange(0, len(X), 0.001)           # Time vector
    y = np.zeros_like(t)                      # Output signal

    # Encode the input signal
    while n < len(X):
        if X[n-1] == 0:
            if X[n] == 0:
                y[(t >= n-1) & (t < n)] = -1
            else:
                y[(t >= n-1) & (t < n)] = -1
        else:
            if X[n] == 0:
                y[(t >= n-1) & (t < n)] = 1
            else:
                y[(t >= n-1) & (t < n)] = 1

        n += 1

    plt.plot(t, y, linewidth=2.5)
    plt.title('Polar RZ line coding')
    plt.grid(True)
    plt.axis([0, len(X), -1.5, 1.5])  # Set plot axis limits
```

```python
# The main function prompts the user to enter a binary input sequence, converts it to a list of integers,
# and then plots the Polar NRZ line coding representation of the input sequence using Matplotlib.



def main():
    print("\n\n***************************************************************************************")
    X = input("---> Enter binary input sequence (e.g., 1011100): ")
    X = [int(bit) for bit in X]
    print("\n\n")
    # Create a new figure and plot the polar NRZ line coding for the input sequence X
    plt.figure()
    plt.subplot(2, 1, 1)
    polarNRZ(X)
    plt.suptitle('Input: ' + ' '.join(map(str, X)))
    plt.show()
    plt.show()

if __name__ == "__main__":
    main()
```
