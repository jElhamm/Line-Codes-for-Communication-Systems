```python
# This code defines a Python function polarRZ that implements Polar Return-to-Zero (RZ) line coding.
# The function takes a binary input sequence X, where it assigns specific voltage levels to represent the binary values.
# If the input bit is 0, the voltage level is set to -1 for that duration.
# For a 1 bit, the voltage level is set to 1 for the first half of the bit period and then drops to 0 for the second half.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def polarRZ(X):
    n = 0
    L = len(X)
    t = np.arange(0, L, 0.001)
    y = np.zeros_like(t)

    while n < L:
        if X[n] == 0:
            y[t >= n] = -1                              # Set values to -1 when the input is 0
        else:
            y[(t >= n) & (t < n + 0.5)] = 1             # Set values to 1 for the first half of the bit period
            y[(t >= n + 0.5) & (t < n + 1)] = 0         # Set values to 0 for the second half of the bit period

        n += 1

    plt.plot(t, y, linewidth=2.5)
    plt.title('Polar RZ line coding')
    plt.grid(True)
    plt.axis([0, len(X), -1.5, 1.5])
```

```python
# The main function prompts the user to enter a binary input sequence, converts it to a list of integers,
# and then plots the Polar RZ line coding representation of the input sequence using Matplotlib.



def main():
    print("\n\n***************************************************************************************")
    X = input("---> Enter binary input sequence (e.g., 1011100): ")
    X = [int(bit) for bit in X]
    print("\n\n")
    # Create a new figure and plot the polar RZ line coding for the input sequence X
    plt.figure()
    plt.subplot(2, 1, 1)
    polarRZ(X)
    plt.suptitle('Input: ' + ' '.join(map(str, X)))
    plt.show()
    plt.show()

if __name__ == "__main__":
    main()
```
