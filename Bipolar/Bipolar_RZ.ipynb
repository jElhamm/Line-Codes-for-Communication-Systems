```python
# This code generates a Bipolar RZ line coding for a binary input sequence X.
# It plots the line coding signal with specific levels for 1 and -1 based on the input data.

# The code loops through the input sequence X and encodes it into the Bipolar RZ format.
# It assigns values of 1 for each '1' input and -1 for each '0' input. It also sets the midpoint value to 0 when there is a transition from '1' to '0'.
# The resultant Bipolar RZ signal is plotted against time with a specific level for each symbol.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bipolarRZ(X):
    n = 0
    L = len(X)
    t = np.arange(0, len(X), 0.001)                               # Time axis
    y = np.zeros_like(t)                                          # Initialize output signal

    while n < len(X)-1:
        if X[n] == 0:
            y[np.where((t >= n) & (t < n+0.5))] = -1              # Assign -1 for '0'
        else:
            y[np.where((t >= n) & (t < n+0.5))] = 1               # Assign 1 for '1'

        # Set midpoint to 0 for transition from '1' to '0'
        if X[n] == 1 and X[n+1] == 0:
            y[int(n+0.5*1000)] = 0

        n += 1

    plt.plot(t, y, linewidth=2.5)                                 # Plot the Bipolar RZ signal
    plt.title('Bipolar RZ line coding')                           # Set plot title
    plt.grid(True)                                                # Display grid lines
    plt.axis([0, len(X)-1, -1.5, 1.5])                            # Set plot axis limits
```

```python
# The main function prompts the user to enter a binary input sequence, converts it to a list of integers,
# and then plots the Polar RZ line coding representation of the input sequence using Matplotlib.



def main():
    print("\n\n***************************************************************************************")
    X = input("---> Enter binary input sequence (e.g., 1011100): ")
    X = [int(bit) for bit in X]
    print("\n\n")
    # Create a new figure and plot the polar RZ line coding for the input sequence X
    plt.figure()
    plt.subplot(2, 1, 1)
    bipolarRZ(X)
    plt.suptitle('Input: ' + ' '.join(map(str, X)))
    plt.show()
    plt.show()

if __name__ == "__main__":
    main()
```
